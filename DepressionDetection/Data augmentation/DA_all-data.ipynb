```python
##### import shutil
from pathlib import Path
from distutils.dir_util import copy_tree
import csv, re, os

target_data_csv = Path('D:/DAIC WOZ # EMNA/full_dataset.csv')
all_data=Path('D:/all_data')
with open(target_data_csv, 'r') as csvFile:
        reader = csv.reader(csvFile)
        for row in reader:
            participant_ID = str(re.split('\t', row[0])[0])
            csv_file_path = 'D:/Emna/DAIC WOZ # EMNA/DAIC WOZ # EMNA/'+participant_ID+'_P/split/Participant/'
            if os.path.exists(csv_file_path):
                csv_file_path2 = 'C:/Users/yobitrust/Desktop/DAIC_augmented_win60/DAIC_augmented/'+participant_ID+'_P/split/Participant/'
                path='D:/all_data/'+participant_ID+'_P/split/Participant/'
                try:  
                    os.makedirs(path)
                except OSError:  
                    print ("Creation of the directory %s failed" % path)
                else:
                    print("Creation of the directory %s success" % path)
                    copy_tree(csv_file_path, path)
                    for filename in os.listdir(csv_file_path2):
                            my_dest ="0" + filename
                            my_source =path + filename
                            my_dest =path + my_dest
                            # rename all the files
                            os.rename(my_source, my_dest)
                    copy_tree(csv_file_path2, path)
```
